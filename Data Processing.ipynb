```python
def split_DNE_dfCAT(r_df_row,back_df):
    dnes = list(map(lambda x: x.strip(),r_df_row.values[0][5].split(';')))
    dnns = list(map(lambda x: x.strip(),r_df_row.values[0][6].split(';')))
    for dne,dnn in zip(dnes,dnns):
        r_df_row['Drug Name Extracted'] = dne
        r_df_row['Drug Name Normalized'] = dnn
        back_df = pd.concat([back_df,r_df_row],axis=0, ignore_index=True)
    return back_df
```

## IOB2 encoding

```python
import nltk
nltk.download('punkt')
```

```python
def process_IOB2(sent,start,end,span):
    tokenize_sents = []
    tags = []
    try:
        e_start = int(start)
        e_end = int(end)
        sents = [sent[:e_start],sent[e_start:e_end],sent[e_end:]]
    except:
        sents = [sent]
    for st in sents:
        tokens = nltk.word_tokenize(st.strip())
        tokenize_sents.extend(tokens)
        if st.lower() == span.lower():
            for idx,t in enumerate(tokens):
                if idx == 0:
                    tags.append('B-')
                else:
                    tags.append('I-')
        else:
            tags.extend(len(tokens) * ['O'])
    
    return tokenize_sents,tags
```

```python
def process_IOB22(rsdf,text,start,end,span):
    tokenize_sents = []
    tags = []
    
    sent = [r for r in rsdf[text]][0]
    el_row = []
    for i in rsdf.index:
        el_row.append((int(rsdf[start][i]),int(rsdf[end][i]),rsdf[span][i]))
    el_row = sorted(el_row , key = lambda x:int(x[1]))
    
    ner_span = [r[2] for r in el_row]
    
    sents = []
    for e_idx,e in enumerate(el_row):
        if e_idx == 0 and e[0] != 0:
            sents.append(sent[0:e[0]])
        sents.append(sent[e[0]:e[1]])
        if e_idx != len(el_row)-1 and sent[el_row[e_idx][1]:el_row[e_idx+1][0]] != ' ':
            sents.append(sent[el_row[e_idx][1]:el_row[e_idx+1][0]])
    if el_row[-1][1] != len(sent):
        sents.append(sent[el_row[-1][1]:])
    
    span_id = 0
    for st in sents:
        tokens = nltk.word_tokenize(st.strip())
        tokenize_sents.extend(tokens)
        if st.lower() == ner_span[span_id].lower():
            if span_id < len(ner_span)-1:
                span_id += 1
            for idx,t in enumerate(tokens):
                if idx == 0:
                    tags.append('B-')
                else:
                    tags.append('I-')
        else:
            tags.extend(len(tokens) * ['O'])
    return tokenize_sents,tags
```

```python
def Process_sentence_tag(df,text,start,end,span,tweet_id):
    p_sentences = []
    p_tags = []
    tweet_idL = []
    for i in df.index:
        if i in pass_list:
            continue
        else:
            if len(df[df[tweet_id] == df[tweet_id][i]]) > 1:
                rs_df = df[df[tweet_id] == df[tweet_id][i]].sort_values([start])
                sentence , tag = process_IOB22(rs_df,text,start,end,span)
            elif len(df[df[tweet_id] == df[tweet_id][i]]) == 1:
                sentence , tag = process_IOB2(df[text][i],df[start][i],df[end][i],df[span][i])
            tweet_idL.append(df[tweet_id][i])
            p_sentences.append(sentence)
            p_tags.append(tag)
    return p_sentences,p_tags   
```

#### Generate Start ID & End ID

```python
def g_startandend(text,tag):
    print(text)
    print(tag)
    cls = False
    drug_id = 0
    st_id,ed_id = 0,0
    for idx,t in enumerate(text.lower()):
        if t == tag.lower()[drug_id] and cls == False and drug_id == 0:
            cls = True
            drug_id += 1
            st_id = idx
        elif t == tag.lower()[drug_id] and cls == True and drug_id != 0:
            drug_id +=1
        elif t != tag.lower()[drug_id] and cls == True and drug_id != len(tag):
            cls = False
            drug_id = 0
        if cls == True and drug_id == len(tag):
            ed_id = idx+1
            break
    return st_id,ed_id
```
